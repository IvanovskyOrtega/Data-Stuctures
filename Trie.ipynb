# Trie

## Definition
A trie is a tree-like data structure wherein the nodes of the tree can store the entire alphabet, and strings/words can be retieved by transversing down a branch path of the tree. (Extracted from [here](https://medium.com/basecs/trying-to-understand-tries-3ec6bede0014))

```python
class Trie():
    head = {}
    
    def add(self,word):
        cur = self.head
        for char in word:
            if char not in cur:
                cur[char] = {}
            cur = cur[char]
        cur['*'] = True
        
    def search(self,word):
        cur = self.head
        for char in word:
            if char not in cur:
                return False
            cur = cur[char]
        return '*' in cur
```

### Let's try the TRIE

```python
trie = Trie()
trie.add('data')
trie.add('structure')
trie.add('computer')
trie.add('science')
trie.add('python')
trie.add('coding')
print(trie.search('code'))
print(trie.search('python'))
print(trie.search('science'))
print(trie.search('data'))
print(trie.search('system'))
```
